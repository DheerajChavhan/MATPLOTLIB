```python
import matplotlib.pyplot as plt
```

```python
# Savefig("filename.extension",dpi=value,bbox_inches="tight")
# (dpi is for clarity of figure and bbox_inches is for cropping)

x=[1,2,3,4]
y=[10,20,15,25]

plt.plot(x,y,color="blue",marker="o")
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.savefig("Line_Plot.pdf",dpi=300,bbox_inches="tight")       # extension can be pdf,png,jpeg,etc.
plt.show()
```
