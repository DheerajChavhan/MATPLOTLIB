```python
import matplotlib.pyplot as plt 
```

```python
product=["A","B","C","D"]
Sales=[1000,750,1500,1300]

plt.bar(product,Sales,color="green",label="Sales 2025" )
plt.xlabel("Product")
plt.ylabel("Sales")
plt.legend()
plt.title("Product Sales Comparison")
plt.ylim(0,1600)
plt.show()
```

```python
product=["A","B","C","D"]
Sales=[1000,750,1500,1300]

plt.barh(product,Sales,color="yellow",label="Sales 2025" )
plt.ylabel("Product")
plt.xlabel("Sales")
plt.legend()
plt.title("Product Sales Comparison")
plt.xlim(0,1600)
plt.show()
```

```python
# plt.pie(values,labels=labels_list,colors=color_list,autopct="%1.1f%%")

regions=["North","South","East","West"]
revenue=[2000,1500,2500,1700]

plt.pie(revenue,labels=regions,autopct="%1.1f%%")
plt.title("Revenue Contribution by Region")
plt.show()
```
