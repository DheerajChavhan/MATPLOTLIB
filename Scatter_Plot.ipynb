```python
import matplotlib.pyplot as plt
```

```python
# plt.scatter(x,y,color="color_name",marker="marker_style",label="label_name")
marks_scored=[40,50,60,70,80,90]
hours_studied=[2,3,4,5,6,7]

plt.scatter(hours_studied,marks_scored,color="Blue",marker="o",label="Student Data")
plt.ylabel("Marks Scored")
plt.xlabel("Hours Studied")
plt.title("Relationship Between Study hours and Exam Score")
plt.legend()
plt.grid()
plt.show()
```

```python
# plt.scatter(x,y,color="color_name",marker="marker_style",label="label_name")

plt.scatter([2,3,4,5,6,7],[40,50,60,70,80,90],color="Blue",marker="o",label="Class A")
plt.scatter([2,3,4,5,6,7],[30,40,50,60,70,80],color="red",marker="o",label="Class B")
plt.ylabel("Marks Scored")
plt.xlabel("Hours Studied")
plt.title("Comparison Between Two Classes")
plt.legend()
plt.grid()
plt.show()
```
